<b> Supervised Regression: Linear Regression Model </b>

```python
import pandas as pd
df = pd.read_csv("data/NBAarchive/Advanced.csv")  # load data set

df
```

```python
#Here I dont want to get into catagorical variables, yet, so I drop all the variables
#that either are catagorical, or things that do not matter like ID valuses.

df = df.drop('player', axis=1)
df = df.drop('birth_year', axis=1)
df = df.drop('lg', axis=1)
df = df.drop('tm', axis=1)
df = df.drop('pos', axis=1)
df = df.drop('seas_id', axis=1)
```

```python
#Run this if you want to drop all columns that contain NaN values.

df=df.dropna(axis=0) #Here i drop NaN values
#df=df.dropna(axis=1) #Here i drop NaN values
#df
```

```python
y = df['bpm']
y
```

```python
X = df.drop('bpm', axis=1)
X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train

print(X_train.shape[0]) #Get count of Training Dataset
print(df.shape[0])
```

```python
X_test

print(X_test.shape[0])
print(df.shape[0])
```

<b> Predictive ML </b>

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
#y_lr_train_pred
#y_lr_test_pred

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
lr_train_mae = mean_absolute_error(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
lr_test_mae = mean_absolute_error(y_test, y_lr_test_pred)
```

```python
import math

print('LR MSE (Train): ', lr_train_mse) #mean squared error
print('LR RMSE (Train): ', math.sqrt(lr_train_mse)) #root mean squared error
print('LR R2 (Train): ', lr_train_r2) #RSquared
print('LR MAE (Train): ', lr_train_mae) #Mean Absolute Error

print(" ") #Spacer

print('LR MSE (Test): ', lr_test_mse)
print('LR RMSE (Test): ', math.sqrt(lr_test_mse))
print('LR R2 (Test): ', lr_test_r2)
print('LR MAE (Test): ', lr_test_mae)
```

MSE: Small = good
RMSE: Small = good
R2: Large/1 = good
MAE: Small = good

Now we can predict NBA player BPM if we have all the other stats to go with it. Which is not useful, because we want variables that can predict the future. In this case, these variables occur at the same time at BPM. 

So we will do something more interesting. I sorted the Excel file of NBA stats by experience (presumably more experience means more data points). Vince Carter has 22 seasons, so I will make two excel files where ~20% are his latter years and ~80% are his former years. 

```python
import pandas as pd
df = pd.read_csv("data/NBAarchive/VinceCarterData.csv")  # load data set

#df
```

```python
df = df.drop('player', axis=1)
df = df.drop('birth_year', axis=1)
df = df.drop('lg', axis=1)
df = df.drop('tm', axis=1)
df = df.drop('pos', axis=1)
df = df.drop('seas_id', axis=1)
```

```python
df=df.dropna(axis=0) #Here i drop NaN values
```

```python
y = df['bpm']
#y
X = df.drop('bpm', axis=1)
#X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle = False, stratify = None)
```

```python
X_train

print(X_train.shape[0]) #Get count of Training Dataset
print(df.shape[0])
```

```python
X_test

print(X_test.shape[0])
print(df.shape[0])
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
#y_lr_train_pred
#y_lr_test_pred

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
lr_train_mae = mean_absolute_error(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
lr_test_mae = mean_absolute_error(y_test, y_lr_test_pred)
```

```python
import math

print('LR MSE (Test): ', lr_test_mse) #mean squared error
print('LR RMSE (Test): ', math.sqrt(lr_test_mse)) #root mean squared error (.2-.5 is good)
print('LR R2 (Test): ', lr_test_r2) #RSquared
print('LR MAE (Test): ', lr_test_mae) #Mean Absolute Error
```

Here we were able to predict Vince Carters efficiency with only ~20 data points.
